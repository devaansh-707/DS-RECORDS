```python
# WEEK 3 DATA SCIENCE LAB
# Similarity and Dissimilarity Measures
# ==========================================

import numpy as np
import pandas as pd
from scipy.spatial import distance
from scipy.stats import spearmanr
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

print("=" * 60)
print("WEEK 3 - SIMILARITY & DISSIMILARITY MEASURES")
print("=" * 60)

# =====================================================
# 1. VECTOR-BASED DISTANCE MEASURES
# =====================================================

print("\n1. VECTOR-BASED DISTANCE MEASURES")
print("-" * 50)

pointA = np.array([2, 4, 6])
pointB = np.array([5, 1, 9])

print("Point A:", pointA)
print("Point B:", pointB)

# Euclidean Distance
euclidean_dist = distance.euclidean(pointA, pointB)
similarity_euclidean = 1 / (1 + euclidean_dist)

print("\nEuclidean Distance:", euclidean_dist)
print("Euclidean Similarity:", similarity_euclidean)

# Manhattan Distance
manhattan_dist = distance.cityblock(pointA, pointB)
similarity_manhattan = 1 / (1 + manhattan_dist)

print("\nManhattan Distance:", manhattan_dist)
print("Manhattan Similarity:", similarity_manhattan)

# Minkowski Distance
minkowski_dist = distance.minkowski(pointA, pointB, p=3)
similarity_minkowski = 1 / (1 + minkowski_dist)

print("\nMinkowski Distance (p=3):", minkowski_dist)
print("Minkowski Similarity:", similarity_minkowski)

# =====================================================
# 2. STATISTICAL MEASURES
# =====================================================

print("\n\n2. STATISTICAL MEASURES")
print("-" * 50)

df = pd.DataFrame({
    'X': [10, 20, 30, 40, 50],
    'Y': [12, 24, 33, 45, 60]
})

print("\nDataset:")
print(df)

# Pearson Correlation
pearson_corr = df.corr(method='pearson')

print("\nPearson Correlation Matrix:")
print(pearson_corr)

# Spearman Correlation
corr_value, p_value = spearmanr(df['X'], df['Y'])

print("\nSpearman Correlation Coefficient:", corr_value)
print("P-Value:", p_value)

# =====================================================
# 3. EDIT-BASED MEASURES
# =====================================================

print("\n\n3. EDIT-BASED MEASURES")
print("-" * 50)

def hamming_distance(str1, str2):

    if len(str1) != len(str2):
        raise ValueError("Strings must be of equal length")

    return sum(ch1 != ch2 for ch1, ch2 in zip(str1, str2))

s1 = "karolin"
s2 = "kathrin"

hamming_dist = hamming_distance(s1, s2)

print("String 1:", s1)
print("String 2:", s2)
print("Hamming Distance:", hamming_dist)

# =====================================================
# 4. TOKEN-BASED MEASURES
# =====================================================

print("\n\n4. TOKEN-BASED MEASURES")
print("-" * 50)

text1 = "data science is fun"
text2 = "science makes data useful"

print("Text 1:", text1)
print("Text 2:", text2)

# Jaccard Index
def jaccard_index(str1, str2):

    set1 = set(str1.split())
    set2 = set(str2.split())

    intersection = set1.intersection(set2)
    union = set1.union(set2)

    return len(intersection) / len(union)

jaccard = jaccard_index(text1, text2)

print("\nJaccard Index:", jaccard)

# Cosine Similarity
vectorizer = CountVectorizer().fit([text1, text2])

vectors = vectorizer.transform([text1, text2])

cos_sim = cosine_similarity(vectors[0], vectors[1])[0][0]

print("Cosine Similarity:", cos_sim)

# =====================================================
# 5. SEQUENCE-BASED MEASURES
# =====================================================

print("\n\n5. SEQUENCE-BASED MEASURES")
print("-" * 50)

def lcs_length(X, Y):

    m = len(X)
    n = len(Y)

    dp = [[0]*(n+1) for _ in range(m+1)]

    for i in range(m):
        for j in range(n):

            if X[i] == Y[j]:
                dp[i+1][j+1] = dp[i][j] + 1

            else:
                dp[i+1][j+1] = max(
                    dp[i][j+1],
                    dp[i+1][j]
                )

    return dp[m][n]

seq1 = "ABCDEF"
seq2 = "AEBDF"

lcs = lcs_length(seq1, seq2)

print("Sequence 1:", seq1)
print("Sequence 2:", seq2)
print("LCS Length:", lcs)

# =====================================================
# FINAL SUMMARY
# =====================================================

print("\n\n" + "=" * 60)
print("SUMMARY")
print("=" * 60)

print("Euclidean Distance :", round(euclidean_dist, 4))
print("Manhattan Distance :", manhattan_dist)
print("Minkowski Distance :", round(minkowski_dist, 4))
print("Pearson Correlation :", round(pearson_corr.iloc[0,1], 4))
print("Spearman Correlation :", round(corr_value, 4))
print("Hamming Distance :", hamming_dist)
print("Jaccard Index :", round(jaccard, 4))
print("Cosine Similarity :", round(cos_sim, 4))
print("LCS Length :", lcs)

print("\nProgram Executed Successfully!")
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age':[25,30,np.nan,40,35],
    'Department':['HR','Finance','Finance',np.nan,'IT']
})

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])

print(df)
```

To identify and remove duplicate records from a dataset.

```python
import pandas as pd

df = pd.DataFrame({
    'ID':[1,2,2,3,4,4],
    'Name':['Alice','Bob','Bob','Charlie','David','David']
})

print(df.drop_duplicates())
```

Experiment 3: Correcting Inconsistent Formats
Aim

To standardize date and text formats.

```python
import pandas as pd

df = pd.DataFrame({
    'Date':['2025-01-05','05/01/2025','Jan 5, 2025']
})

df['Date'] = pd.to_datetime(df['Date'], format='mixed', dayfirst=True).dt.strftime('%Y-%m-%d')

print(df)
```

Experiment 4: Detecting Outliers
Aim

To identify outliers using the IQR method.

```python
import pandas as pd

data = [10,12,15,18,20,22,100]

df = pd.DataFrame(data, columns=['Values'])

Q1 = df['Values'].quantile(0.25)
Q3 = df['Values'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[(df['Values'] < lower) | (df['Values'] > upper)]

print(outliers)
```

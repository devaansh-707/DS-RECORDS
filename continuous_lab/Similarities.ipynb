```python
import numpy as np
from scipy.spatial import distance
point1=np.array([2,4,6])
point2=np.array([5,1,9])
euclidean_dist=distance.euclidean(point1,point2)
print("Euclidean distance",euclidean_dist)
similarity_euclidean=1/(1+euclidean_dist)
print("euclidean similarity",similarity_euclidean)

manhattan=distance.cityblock(point1,point2)
print("Manhattan distance",manhattan)
similarity_manhattan=1/(1+manhattan)
print("Manhattan similarity",similarity_manhattan)

# mikowski distance with p=3
minkowski=distance.minkowski(point1,point2,p=3)
print("Minkowski distance",minkowski)
similarity_minkowski=1/(1+minkowski)
print("Minkowski similarity",similarity_minkowski)
# mikowski distance with p=2...it is equal to euclidean
minkowski=distance.minkowski(point1,point2,p=2)
print("Minkowski distance",minkowski)
similarity_minkowski=1/(1+minkowski)
print("Minkowski similarity",similarity_minkowski)
```

```python
# hamming distance between strings
def hamming_dist(str1,str2):
  if len(str1)!=len(str2):
    raise ValueError('Strings must be of equal length')
  return sum(ch1!=ch2 for ch1,ch2 in zip(str1,str2))
s1='abc'
s2='acb'
dist=hamming_dist(s1,s2)
print(dist)
```

```python
# sets/text jaccard index
def jaccard_index(str1,str2):
  s1,s2=set(str1.split()),set(str2.split())
  intersection=s1.intersection(s2)
  union=s1.union(s2)
  return len(intersection)/len(union)

c1="data science"
c2="science is fun"
print("jaccard index",jaccard_index(c1,c2))

#  cosine similarity
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer=CountVectorizer().fit([c1,c2])
vectors=vectorizer.transform([c1,c2])
cos_sim=cosine_similarity(vectors[0],vectors[1])[0][0]
print("cosine similarity",cos_sim)

```

```python
# sequence based measures:longest common subsequence
def lcs_length(X,Y):
  m,n=len(X),len(Y)
  dp=[[0]*(n+1) for _ in range(m+1)]

  for i in range(m):
    for j in range(n):
      if X[i]==Y[j]:
        dp[i+1][j+1]=dp[i][j]+1
      else:
        dp[i+1][j+1]=max(dp[i][j+1],dp[i+1][j])
  return dp[m][n]
s1="ABCDEF"
s2="AEBCDF"
length=lcs_length(s1,s2)
print("length of longest common subsequence",length)
```

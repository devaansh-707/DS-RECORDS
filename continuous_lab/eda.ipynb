```python
import matplotlib.pyplot as plt
y1=[]
y2=[]
x=range(-100,100,10)
for i in x:
  y1.append(i**2)
for i in x:
  y2.append(-i**2)
plt.plot(x,y1)
plt.plot(x,y2)
# plt.show()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('simple graph')
plt.axhline()
plt.axvline()
plt.show()
```

```python
# skewness
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
print("First 5 rows of the dataset:")
print(df.head())
skewness_values=df.skew()
print("\nskewness values:")
print(skewness_values)
kurtosis_values=df.kurt()
print("\nkurtosis values:")
print(kurtosis_values)
```

```python
import pandas as pd
tips=pd.read_csv('tips.csv')

print("First 5 rows of the dataset:")
print(tips.head())
skewness_values=tips._get_numeric_data().skew()
print("\nskewness values:")
print(skewness_values)
kurtosis_values=tips._get_numeric_data().kurt()
print("\nkurtosis values:")
print(kurtosis_values)
```
